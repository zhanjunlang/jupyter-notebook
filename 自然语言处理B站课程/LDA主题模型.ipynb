## 什么是LDA
Latent Dirichlet Allocation<br>
是一种无监督的贝叶斯模型，是一种典型的词袋模型。<br>
核心公式：
$$P(w|d)=P(w|t)P(t|d)$$
* w：出现在给定文档中的一个单词
* d：给定文档
* t：某个主题

## 公式推导
gamma函数<br>
多项分布<br>
beta分布<br>
狄利克雷分布（beta分布在高纬度上的推广）<br>

Unigram model<br>
$d=(w_1,w_2,...,w_N)$ <br>
用$p(w_n)$表示词$w_n$的先验概率，则生成文档d的概率为<br>
$p(d)=\prod_{n=1}^{N}p(w_n)$

Mixture of unigram model<br>
设有k个主题$(z_1,...,z_k)$,<br>
$p(d)=p(z_1)\prod_{n=1}^{N}p(w_n|z_1)+...+p(z_k)\prod_{n=1}^{N}p(w_n|z_k)$<br>
$=\sum_{z}p(z)\prod_{n=1}^{N}p(w_n|z)$<br>

PLSA model<br>
一篇文章可能有多个主题，可以看做有两个骰子：第一个是“文档-主题”骰子，第二个是“主题-单词”骰子。模型的过程是先扔第一个骰子，确定一个模型，然后根据选择的模型扔第二个骰子生成一个单词。反复重复两个过程，生成一堆单词，然后挑选符合给定文章单词分布的。

LDA model<br>
在PLSA model的基础上，在两个骰子的概率每次扔都是不一样的，比如第一个骰子，它的不同主题的概率在PLSA中是确定的，但在LDA中每次扔的时候每个主题的概率是根据狄利克雷分布生成的。
